```python
from fastai.tabular.all import *
```

```python
path = Path('../../data/kaggle-wine/')
```

```python
df = pd.read_csv(path/'winequality-red.csv')
```

```python
df.loc[:,'quality'].unique(), df.columns
```

```python
cont_names = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
        'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
        'pH', 'sulphates', 'alcohol']
```

```python
df.head()
```

```python
splits = RandomSplitter(valid_pct=0.2)(range_of(df))
```

```python
dep_var = 'quality'
```

```python
to = TabularPandas(df, procs=[Categorify, FillMissing,Normalize],
                   cont_names = cont_names,
                   y_names=dep_var,
                   splits=splits,y_block = CategoryBlock())
```

```python
dls = to.dataloaders(bs=64)
```

```python
dls.show_batch()
```

```python
learn = tabular_learner(dls, metrics=accuracy)
```

```python
learn.fit_one_cycle(10)
```

```python
learn.show_results()
```
